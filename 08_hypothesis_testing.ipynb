# 1. Import libraries

```python
import pandas as pd
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
sns.set(style="whitegrid")
```

# 2. Load dataset from files upload



```python
from google.colab import files

uploaded = files.upload()

# Assuming single file upload
for filename in uploaded.keys():
    data = pd.read_csv(filename)
    print(f"Loaded {filename} successfully.")

data.head()
```

#3. One Sample t-test

```python
# Replace 'col_name' with your column of interest
col_name = 'Age'  # Example column

# H0: mean = 30
# Ha: mean != 30
t_stat, p_val = stats.ttest_1samp(data[col_name], 30)

print(f"One Sample t-test for {col_name} vs 30")
print("t-statistic:", t_stat)
print("p-value:", p_val)

if p_val < 0.05:
    print("Reject Null Hypothesis: Significant difference from 30")
else:
    print("Fail to Reject Null Hypothesis: No significant difference from 30")
```

#4. Two Sample Independent t-test

```python
# Replace 'gender_col' with your categorical column name and 'num_col' with numerical column
gender_col = 'Gender'
num_col = 'Age'

# Check unique groups
print(data[gender_col].unique())

# Perform t-test between two groups (e.g., Male and Female)
group1 = data[data[gender_col]=='Male'][num_col]
group2 = data[data[gender_col]=='Female'][num_col]

t_stat, p_val = stats.ttest_ind(group1, group2, equal_var=False)

print(f"Two Sample t-test for {num_col} by {gender_col}")
print("t-statistic:", t_stat)
print("p-value:", p_val)

if p_val < 0.05:
    print("Reject Null Hypothesis: Significant difference between groups")
else:
    print("Fail to Reject Null Hypothesis: No significant difference between groups")
```

#5. Chi-Square Test of Independence

```python
# Replace with actual categorical columns in your data
cat1 = 'Gender'
cat2 = 'family_history'

# Create contingency table
cont_table = pd.crosstab(data[cat1], data[cat2])

# Perform chi-square test
chi2, p, dof, expected = stats.chi2_contingency(cont_table)

print(f"Chi-Square Test between {cat1} and {cat2}")
print("Chi-square statistic:", chi2)
print("p-value:", p)

if p < 0.05:
    print("Reject Null Hypothesis: Variables are dependent")
else:
    print("Fail to Reject Null Hypothesis: Variables are independent")
```
