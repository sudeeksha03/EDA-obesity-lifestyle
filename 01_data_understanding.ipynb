# **Data understanding**

```python
# Step 1: Upload CSV file (Google Colab)
from google.colab import files
import pandas as pd

# Upload the file manually
uploaded = files.upload()

# Load the uploaded file into a DataFrame
for filename in uploaded:
    df = pd.read_csv(filename)
    print(f"âœ… File '{filename}' loaded successfully.\n")

# Step 2: Basic Data Preview
print("ðŸ”¹ First 5 rows:")
display(df.head())

# Step 3: Structure and Metadata
print("\nðŸ”¹ DataFrame Info:")
df.info()

# Step 4: Shape and Columns
print("\nðŸ”¹ Dataset Shape:", df.shape)
print("ðŸ”¹ Column Names:", df.columns.tolist())

# Step 5: Data Types
print("\nðŸ”¹ Data Types:")
print(df.dtypes)

# Step 6: Unique values in each column
print("\nðŸ”¹ Unique Values per Column:")
print(df.nunique())

# Step 7: Random Sample Rows
print("\nðŸ”¹ Random Sample Rows:")
display(df.sample(5))

# Step 8: Missing Values
print("\nðŸ”¹ Missing Values:")
print(df.isnull().sum())
```
