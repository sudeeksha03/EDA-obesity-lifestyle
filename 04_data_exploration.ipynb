Dataset Overview

```python
print("üìè Dataset Shape:", df.shape)
print("\nüßæ Data Types:")
print(df.dtypes)

print("\nüîÅ Check for Null Values:")
print(df.isnull().sum())
```

Distribution of Target Variable

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 5))
sns.countplot(x='Obesity', data=df)
plt.title("üéØ Distribution of Obesity Levels (Target)")
plt.xticks(rotation=45)
plt.show()
```

Correlation Matrix

```python
plt.figure(figsize=(14, 10))
sns.heatmap(df.select_dtypes(include='number').corr(), annot=True, cmap='coolwarm', fmt='.2f')
plt.title("üìä Correlation Heatmap")
plt.show()
```

Univariate Analysis

```python
numerical_cols = df.select_dtypes(include=['int64', 'float64']).columns.tolist()
for col in numerical_cols:
    plt.figure(figsize=(6, 4))
    sns.histplot(df[col], kde=True, bins=30)
    plt.title(f"üìâ Distribution of {col}")
    plt.show()
```

Categorical Feature Distributions

```python
categorical_cols = df.select_dtypes(include='int64').columns.tolist()
categorical_cols = [col for col in categorical_cols if col != 'NObeyesdad']

for col in categorical_cols:
    plt.figure(figsize=(6, 4))
    sns.countplot(x=col, data=df)
    plt.title(f"üì¶ Count Plot: {col}")
    plt.xticks(rotation=45)
    plt.show()
```

# Bivariate Analysis

BMI vs Obesity Level

```python
# Calculate BMI
df['BMI'] = df['Weight'] / (df['Height']**2)

plt.figure(figsize=(10, 5))
sns.boxplot(x='Obesity', y='BMI', data=df)
plt.title("üìå BMI Distribution by Obesity Level")
plt.xticks(rotation=45)
plt.show()
```
